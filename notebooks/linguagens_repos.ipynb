### Autentica√ß√£o

##### Solicita√ß√µes autenticadas t√™m um limite de taxa mais alto. Quando um usu√°rio faz uma solicita√ß√£o autenticada, ele fornece credenciais que comprovam sua identidade, o que permite que a API confine nele e lhe conceda acesso a recursos e funcionalidades adicionais.

##### Al√©m disso, a maioria das APIs estabelece limites para o n√∫mero de solicita√ß√µes que o usu√°rio pode fazer em um determinado per√≠odo de tempo, conhecido como limite de taxa (rate limit). Esse mecanismo ajuda a proteger os servidores contra sobrecarga, uso abusivo ou ataques, garantindo que os recursos da API sejam distribu√≠dos de forma justa entre todos os usu√°rios.

##### Quando um usu√°rio faz solicita√ß√µes autenticadas, a API geralmente  permite que ele fa√ßa mais solicita√ß√µes em um determinado per√≠odo de tempo, devido √† maior confian√ßa e credibilidade que a autentica√ß√£o fornece.

#### Paginando os reposit√≥rios

```python
'''
Aguardamos um pouco para tudo ser executado. Em seguida, voltaremos 
√† c√©lula onde definimos a URL com todos os endpoints que utilizamos 
para extrair dados dos reposit√≥rios. Nessa c√©lula, copiaremos as 
vari√°veis api_base_url, owner e url e colaremos em uma c√©lula mais
abaixo, na se√ß√£o Paginando os reposit√≥rios, apenas para termos um 
acesso mais f√°cil a elas.
'''

api_base_url = 'https://api.github.com'
owner = 'amzn' # username de quem vamos extrair os dados
url = f'{api_base_url}/users/{owner}/repos'

url
```

```python
'''
Na primeira linha, estamos definindo uma lista vazia com o nome de 
repos_list. Essa lista ser√° utilizada para armazenar cada uma das 
p√°ginas de onde extrairemos os dados com seus respectivos 
reposit√≥rios.
Logo abaixo dessa defini√ß√£o da lista, temos um la√ßo for. 
Precisaremos desse la√ßo porque, como sabemos, h√° cinco p√°ginas que 
precisamos percorrer.
'''

repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)

```

```python
'''
Analisando o conte√∫do da nossa lista, notaremos que ela √© uma lista de 
sublistas, e cada uma dessas sublistas √© composta por v√°rios 
dicion√°rios. Cada sublista existente na repos_list corresponde a 
uma das p√°ginas de reposit√≥rios do GitHub da Amazon, e cada
dicion√°rio existente nessas sublistas s√£o os reposit√≥rios existentes
em cada uma das p√°ginas. 
'''

repos_list
```

```python
''' 
Na c√©lula seguinte, se escrevermos len(repos_list) para verificarmos
o tamanho da nossa lista, conferimos que ela tem o tamanho 5, 
que √© a mesma quantidade de p√°ginas de reposit√≥rios que temos no 
GitHub da Amazon.

'''
# Tamanho de SUBLISTAS = Qtde de p√°ginas de repositorios na pagina da amazon
len(repos_list)
```

```python
# Tamanho da primeira p√°ginas que temos
len(repos_list[0])
```

### Transformando os dados

##### Nome dos reposit√≥rios

```python
''' 
Uma vez que selecionamos a p√°gina, podemos selecionar o reposit√≥rio. 
Ent√£o, basta incluir mais um par de colchetes e colocar o √≠ndice do 
reposit√≥rio
'''
repos_list[0][2]['name']
```

```python
''' 
Agora, vamos percorrer toda a lista repos_list para pegar apenas os 
nomes e salv√°-los em uma lista.
'''
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```
